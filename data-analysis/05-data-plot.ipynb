2D绘图

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

#一维数据集

```python
np.random.seed(1000)
y=np.random.standard_normal(20)
y
```

```python
x=range(len(y))
plt.plot(x,y)
```

#如果只提供y值，plot会以索引值作为对应的x值

```python
plt.plot(y)
```

```python
plt.plot(y.cumsum())
```

```python
plt.plot(y.cumsum())
#增加网格
plt.grid(True)
#调节轴的范围
plt.axis('tight')
```

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.xlim(-1,20)
plt.ylim(np.min(y.cumsum())-1,np.max(y.cumsum())+1)
```

```python
y.cumsum()
```

```python
plt.figure(figsize=(7,4))
plt.plot(y.cumsum(),'b',lw=1.5)
plt.plot(y.cumsum(),'r*')
plt.plot(y.cumsum(),'yo')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```
